## Logistic Regression For Multiclass Classification

```python
# make a prediction with a multinomial logistic regression model
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
# define dataset
X, y = make_classification(n_samples=1000, n_features=10, n_informative=5, n_redundant=5, n_classes=3, random_state=1)
```

```python

# define the multinomial logistic regression model For multiclass classification
model=LogisticRegression(solver='lbfgs', max_iter=1000)
#  in newer versions of scikit-learn, the 'lbfgs' solver is the default and supports multinomial logistic regression.
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# trianing model
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
model.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

#### Implement Logistic Regression on the Iris Dataset to classify the species of iris flowers based on their features.


```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
dataset=load_iris()
```

```python
dataset.keys()
```

```python
x=pd.DataFrame(dataset.data,columns=dataset.feature_names)
y=pd.DataFrame(dataset.target,columns=['target'])
```

```python
x
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model=LogisticRegression(solver='lbfgs', max_iter=1000)
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
model.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

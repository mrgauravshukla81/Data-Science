# PERFORMANCE MARTRICS FOR REGRESSION


1.R squared:
=====

R square = 1 - (SSres / SStot)

Where:
- SSres is the sum of squares of residuals (the differences between observed and predicted values).
- SStot is the total sum of squares (the differences between observed values and the mean of observed values).

R squared is between 0 and 1.

-if we add  highly correlated features to predict target then R squared value will be increased with higher value.

-However -if we add  less correlated features to predict target then R squared value will be increased with lower value.


 ## 2.Adjusted R squared:


Adjusted R squared = 1 - [(1 - R squared) * (n - 1) / (n - p - 1)]

Where:
- n is the number of observations.
- p is the number of predictors (independent variables) in the model.   

R squared >> Adjusted R squared
ex:

>p=2 R squared = 0.80==Adjusted R squared = 0.75

> p=3 R squared 85=== Adjusted R squared = 78

> p=4 R squared 88=== Adjusted R squared = 76 =>independent variable is not significant to predict target

DIFFERENT TYPES OF COST FUNCTIONS
============================

### Mean Squared Error (MSE):
MSE = (1/n) * Σ(actual - predicted)² =>Quadratic Equation

advantages:
- equation is differentiable
- it has only one local or global minimum

disadvantages:
- sensitive to outliers
- it does not provide error in the same unit as the target variable

### Mean Absolute Error (MAE):
MAE = (1/n) * Σ |yi - ŷi|
Where:
- yi represents the actual values.
- ŷi represents the predicted values.
- n is the number of observations.

advantages:
- MAE is less sensitive to outliers compared to MSE.
- It provides error in the same unit as the target variable.

disadvantages:
- convergence can be slower compared to MSE.

### Root Mean Squared Error (RMSE):
RMSE = sqrt((1/n) * Σ(actual - predicted)²)

advantages:
- RMSE provides error in the same unit as the target variable.
- differentiable

disadvantages:
- sensitive to outliers

```python
import matplotlib.pyplot as plt 
```

```python
import numpy as np
```

```python
x=np.linspace(0,10,200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.xlabel('X axis label')
plt.ylabel('Y axis label')
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
plt.scatter(x,y)
```

```python
color=np.random.rand(50)
size=np.random.rand(50)
plt.scatter(x,y,marker='*',c=color,s=size*1000,alpha=0.5,cmap='viridis')   
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter Plot Example')
```

```python
x=['x1','x2','x3','x4']
y=np.random.rand(4)
plt.bar(x,y,color=['red','blue','green','orange'])
plt.xlabel('Categories')
plt.ylabel('Values')
```

```python
plt.barh(x,y,color=['red','blue','green','orange'])
```

```python

plt.figure(figsize=(4,2))#to control size of plot
plt.grid()#to add grid
plt.plot(x,y,'--r')
plt.show()
```

```python
data=[1,2,3,4,3,1,2,3,4,5,6,5,4,3,2,1]
plt.hist(data,color='purple',alpha=0.7)
plt.show()
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
z=np.random.rand(50)
fig=plt.figure()
ax=fig.add_subplot(projection='3d')
ax.scatter(x,y,z,s=50)
plt.show()
```

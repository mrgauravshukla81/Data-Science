```python
import seaborn as sns
```

```python
iris=sns.load_dataset('iris')
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=iris)
```

```python
#for histogram
sns.displot(iris['sepal_length'])
```

```python
tips=sns.load_dataset('tips')
#hotel ka data hai
```

```python
tips
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip)
```

```python
tips['smoker'].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='smoker')
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,hue='size')
```

```python
tips['size'].value_counts()
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size')
#hue color dega aur style shape dega
```

```python
sns.scatterplot(x=tips.total_bill,y=tips.tip,data=tips,style='size',hue='time')
#do parameter ek saath use kar sakte hai
```

```python
sns.catplot(x='day',y='total_bill',data=tips)
#kis din zyada bill hota hai
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
#beech me scatter plot aur side me histogram
```

```python
sns.pairplot(tips)
#saare columns ke beech me relation dikhata hai
```

```python
sns.jointplot(iris)
```

Data Interpolation
Data interpolation is the process of estimating unknown values within a dataset based on the known values. In Python, there are various libraries available that can be used for data interpolation, such as NumPy, SciPy, and Pandas. Here is an example of how to perform data interpolation using the NumPy library:

1.Linear interpolation

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1, 2, 3,4,5])
y=np.array([2,4,6,8,10])
plt.scatter(x,y)
```

```python
#interpolate the data using interpolation
x_new=np.linspace(1,5,10)#create new x values
y_interp=np.interp(x_new,x,y)#interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp)
```

Cubic Interpolation With scipy

```python
x=np.array([1, 2, 3,4,5])
y=np.array([1,8,27,64,125])
```

```python
from scipy.interpolate import interp1d
```

```python
#create a cubic interpolation function
f=interp1d(x,y,kind='cubic')
```

```python
#interpolate the data using interpolation
x_new=np.linspace(1,5,10)#create new x values
y_interp=f(x_new)#interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp, color='red')
```

Polynomial Interpolation

```python
x=np.array([1, 2, 3,4,5])
y=np.array([1,4,9,16,25]) 
```

```python
#interpolate the data using polynomial interpolation
p=np.polyfit(x,y,2)#fit a polynomial of degree 2
```

```python
x_new=np.linspace(1,5,10)#create new x values
y_interp=np.polyval(p,x_new)#interpolate y values
print(y_interp)
```

```python
plt.scatter(x_new,y_interp, color='green')
```

Handling Outliers

# 5 number Summary
1.Minimum Value
2. Q1-25 Percentile
3. Median-50 Percentile
4. Q3-75 Percentile
5. Maximum Value

```python
import numpy as np
lst_marks=[45,32,67,89,90,78,56,49,73,88,1000,1100]
```

```python
Q1=np.percentile(lst_marks,[25])
print(Q1)
```

```python
minimum,Q1,median,Q3,maximum=np.percentile(lst_marks,[0,25,50,75,100])
print(minimum,Q1,median,Q3,maximum)
```

```python
IQR=Q3-Q1
lower_fence=Q1-1.5*IQR
upper_fence=Q3+1.5*IQR
```

```python
lower_fence,upper_fence
```

```python
outliers=[]
for i in lst_marks:
    if i<lower_fence or i>upper_fence:
        outliers.append(i)
print(outliers)
```

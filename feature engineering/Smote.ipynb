SMOTE(synthetic Minority Oversampling Technique)

```python
from sklearn.datasets import make_classification
```

```python
#x is independent features and y is dependent feature
x,y=make_classification(n_samples=1000, n_features=2,n_redundant=0,n_clusters_per_class=1,weights=[0.9],random_state=1)
```

```python
x.shape
```

```python
import pandas as pd
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
df=pd.concat([df1,df2],axis=1)
```

```python
df
```

```python
df['target'].value_counts()##this is imbalanced data
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['f1'],df['f2'],c=df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
#transforming the data
oversample=SMOTE()
x,y=oversample.fit_resample(df[['f1','f2']],df['target'])
```

```python
x.shape,y.shape
```

```python
len(y[y==0])
```

```python
df1=pd.DataFrame(x,columns=['f1','f2'])
df2=pd.DataFrame(y,columns=['target'])
oversample_df=pd.concat([df1,df2],axis=1)
```

```python
plt.scatter(oversample_df['f1'],oversample_df['f2'],c=oversample_df['target'])
```

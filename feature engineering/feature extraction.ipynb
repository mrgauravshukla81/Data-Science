1.Standardisation---

Z-score=x-mean/std

```python
import seaborn as sns
```

```python
df=sns.load_dataset('tips')
```

```python
total_bill=list(df['total_bill'])
```

```python
import numpy as np
mean=np.mean(total_bill)
std=np.std(total_bill)
```

```python
mean,std
```

```python
normalised_data=[]
for i in total_bill:
    z_score=(i-mean)/std
    normalised_data.append(z_score)
print(normalised_data)
```

```python
sns.histplot(total_bill)
```

```python
sns.histplot(normalised_data)#dono graph same dikhenge bs x axis ka scale alag hoga
```

```python
#directly bhi hm standardisation kr skte h sklearn se
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
Scaled_data=scaler.fit(df[['total_bill','tip']])
```

```python
import pandas as pd
pd.DataFrame(scaled_data.transform(df[['total_bill','tip']]), columns=['total_bill','tip'])
```

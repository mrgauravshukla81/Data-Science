## Ridge Regreesion (L2 regularisation)
- for reducing overfitting

- new cost function

cost function = (1/n) * Σ |yi - ŷi| + lembda *  Σ(slope)^2
- where lembda is > HYPERPARAMETER

### Effect of λ (lambda)

λ = 0 → Same as normal linear regression

Small λ → Slight regularization

Large λ → Strong regularization → coefficients shrink close to zero

Important: Ridge regression shrinks coefficients, but never makes them exactly zero.
